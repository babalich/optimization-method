<b>2. Пусть есть компания, занимающаяся доставкой Belivery Club.
Есть N зон, в которых есть спрос на курьеров в каждый час. Внутри дня в разные часы спрос различен. Каждой зоне i в час h необходимо необходимо n_{hi} курьеров. Для каждой зоны есть штраф из-за недостаточно 1 курьера - s_i. Так же у каждого курьера есть предпочтения относительно часов и зон работы. Необходимо распределить M курьеров таким образом, чтобы суммарный штраф был минимальный.
(описать формулу в общем случае, т.е. без каких-либо числовых данных)</b>

Пронумеруем курьеров:
пусть $x_{ih}^k$ = 1, если k-курьер работает в час h в i зоне

Тогда в одной зоне в час h работает $$\sum_{k=1}^{M} x_{ih}^k$$

Штраф в час в этой зоне
$$ f_{ih} = \begin{cases} 0, \space если \space \sum_{k=1}^{M} x_{ih}^k - n_{ih} >0 \\
s_i * (\sum_{k=1}^{M} x_{ih}^k - n_{ih}), в\spaceпротивном\spaceслучае. \end {cases}$$ 



Штраф за день $$\sum_{h=1}^{24}f_{ih}$$

Штраф по всем зонам в течении дня:


$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih}$$

Далее нам нужно учесть предпочтения сотрудников. Нельзя рассматривать предпочтения как жесткие ограничения, поскольку мы можем попасть в недопустимую ситуацию, например: 1) никто не хочет доставлять еду в i-ый район; 2) никто не хочет работать с 4 до 6 утра. <br>
Нам нужно обеспечить круглосуточную доставку во все зоны, следовательно задача учета предпочтения курьеров так же является задачей оптимизации. Решим их по отдельности.

Пусть $z_{ih}^k$ = 1, если k-курьер работает в i-зоне, при этом h-час для него является нежелательным <br>
$\zeta_h$ - условный штраф модели за h-час (в разные часы штраф может быть разным),<br> тогда штраф за "неудобную работу" в i-зоне h-час:
$$\zeta_h * \sum_{k=1}^{M} z_{ih}^k$$

Тогда общий штраф по всем зонам за день будет равен: $$ \sum_{i=1}^{N} \sum_{h=1}^{24}(\zeta_h * \sum_{k=1}^{M} z_{ih}^k)$$ 

Аналогично для неудобных зон: <br>
Пусть $y_{ih}^k$ = 1, если k-курьер работает в h-час, при этом i-зона ему не нравится <br>
$\gamma_i$ -  штраф за назначение к-курьера в "плохую" i-зону <br>
тогда штраф по i зоне и h-час будет равен: 
$$\gamma_i * \sum_{k=1}^{M} y_{ih}^k$$

Тогда общий штраф по всем зонам за день будет равен: $$ \sum_{i=1}^{N} \sum_{h=1}^{24}(\gamma_i * \sum_{k=1}^{M} y_{ih}^k)$$ 

Соответственно, целевая функция примет вид:
$$(\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih}) + \sum_{i=1}^{N} \sum_{h=1}^{24}\sum_{k=1}^{M}(\gamma_i * y_{ih}^k + \zeta_h *  z_{ih}^k) \rightarrow min $$
